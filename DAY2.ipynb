```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
from sklearn import cluster
np.random.seed(42)
cluster1 = np.random.normal(loc=[2,2],scale=[0.6],size=(100,2))
cluster2 = np.random.normal(loc=[7,7],scale=[0.6],size=(100,2))
cluster3 = np.random.normal(loc=[2,7],scale=[0.6],size=(100,2))
X=np.vstack([cluster1,cluster2,cluster3])
```

```python
plt.figure(figsize=(7,6))
plt.scatter(X[:,0],X[:,1])
plt.title("Original Data(Bina label ka data)")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.show()
```

```python
inertia_value = []
k_range  = range(1,11)
for k in k_range:
  kmeans = KMeans(n_clusters=k,random_state=10).fit(X)
  inertia_value.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(7,6))
plt.plot(k_range,inertia_value,marker="o")
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia Value")
plt.grid(True)
plt.show()
```

```python
optimal_k=3
kmeans = KMeans(n_clusters=optimal_k,random_state=10)

labels = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(7,6))
plt.scatter(X[:,0],X[:,1],c=labels, cmap= 'viridis')
plt.scatter(centroids[:,0], centroids[:,1], marker= 'x', s=250, linewidths= 3, color= 'red', label= 'centroid')
plt.title("clustered data k=3 ab hamne clustring kar di h")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.legend()
plt.grid(True)
plt.show()
```
