```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
data = {
    "hours" : [2.5,5.1,3.2,8.5,3.5,1.5,9.2],
    "marks" : [50,85,74,95,87,40,99]
 }
```

```python
df = pd.DataFrame(data)
df.shape
```

```python
df.describe()
```

```python
plt.figure(figsize=(7,2))
sns.scatterplot(data=df, x="hours",y="marks")
plt.title("Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python
x = df.drop("marks",axis=1)

y = df["marks"]

print(x.head)
print(y.head)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print("Trainig Samples:",x_train.shape,y_train.shape)
print("Testing Samples:",x_test.shape,y_test.shape)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
coeff_df = pd.DataFrame(model.coef_,x.columns,columns=["Coefficient"])
coeff_df
```

```python
y_pred = model.predict(x_test)

comparision = pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
comparision
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Actual Price")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Price")
plt.show()
```

```python
#New student details
new_student = pd.DataFrame({
    "hours" : [0.5]


})

predicted_price = model.predict(new_student)
print("Predicted marks:", predicted_price[0])
```
